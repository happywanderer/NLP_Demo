```python
def extract_ner(text):
    doc1 = [(ent.text, ent.label_) 
            for ent in nlp(text).ents 
            if ent.label_ not in ['CARDINAL', 'ORDINAL', 'QUANTITY', 'MONEY', 'PERCENT', 'TIME']
           ]
    
    print([(ent.text, ent.label_) for ent in nlp(text).ents])
    
    doc2 = list(set(doc1))
    
    dict_1=dict() 
  
    for a,b in doc2: 
        dict_1.setdefault(b, []).append(a) 
    
    return dict_1
```

```python
def add_ner_df(df):
    df['metadata'] = df.text.apply(extract_ner)
```

```python
%%time
add_ner_df(df)
```

```python
def do_vis(text):
    doc = nlp(text)
    displacy.render(doc, style="ent", page=True)
    sentence_spans = list(doc.sents)
    displacy.render(sentence_spans, style="dep", page=True)
    
```
